```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: 5 * x ** 2 + x + 1
interval = np.array([-3, 0])
n = 6
```

## Метод прямоугольников

```python
h = (interval[1] - interval[0]) / n
h
```

```python
intervals = np.array([interval[0] + i* h for i in range(n + 1)])
intervals
```

```python
I = h * np.sum(f(intervals[:-1]))
I
```

```python
x = np.linspace(-3, 0, 1000)
y = f(x)
plt.plot(x, y, color='black');
plt.bar(intervals[:-1], f(intervals[:-1]), alpha=0.75, align='edge', width=0.5, color='gray', edgecolor='black', linewidth=0.5);
```

## Метод трапеций

```python
I = (h / 2) * np.sum([f(intervals[i]) + f(intervals[i + 1]) for i in range(n)])
I
```

```python
x = np.linspace(-3, 0, 1000)
y = f(x)
plt.plot(x, y, color='black');
plt.plot(intervals, f(intervals))
for i in range(n):
    plt.vlines(intervals[i], 0, f(intervals)[i], edgecolor='black', linewidth=0.5)
```

## Метод парабол (Симпсона)

```python
intervals = []
intervals = np.array([
    [interval[0] + i* h, interval[0] + (i + 1)* h, interval[0] + (i + 2)* h] for i in range(0, n, 2)
])
intervals
```

```python
I = (h / 3) * np.sum([
    f(i[0]) + 4 * f(i[1]) + f(i[2]) for i in intervals
])
I
```
