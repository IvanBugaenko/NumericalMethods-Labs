```python
import numpy as np
import matplotlib.pyplot as plt
```

Изучается случайная величина $X$ - число выпавших очков при бросании кубика. Его подбросили 60 раз. Требуется провести первичную обработку статистических данных:
1. записать вариационный ряд;
2. составить статистический ряд;
3. построить полигон частот;
4. найти и построить график эмпирической функции распределения;
5. найти характеристики выборки: выборочную среднюю, выборочную дисперсию и исправленную выборочную дисперсию, стандарт, размах выборки, моду, асимметрию и эксцесс.

```python
data = np.array([
    1, 2, 5, 3, 3, 1, 2, 6, 4, 6, 3, 6, 4, 2, 1, 5, 3, 1, 6, 4,
    5, 4, 2, 2, 4, 2, 6, 6, 4, 4, 2, 1, 6, 6, 4, 2, 5, 4, 1, 6,
    4, 1, 5, 6, 3, 2, 4, 4, 1, 1, 5, 6, 4, 6, 5, 4, 1, 4, 5, 3
])
```

# Запись вариационного ряда

```python
var_row = np.array(sorted(data))
print('Вариационный ряд:', *var_row)
```

# Запись статистического ряда

```python
stat_row, n_row = np.unique(data, return_counts=True)
print('Статистический ряд:', *stat_row)
```

# Полигон относительных частот

```python
p_row = n_row / len(data)
plt.plot(stat_row, p_row)

plt.title('Полигон относительных частот')
plt.xlabel("Значения СВ $X$")
plt.ylabel("Относительная частота значений В в ряду")
plt.show()
```

# Эмпирическая функция распределения

Построим таблицу с характеристиками ряда:
| $x_i$ | Частота $n_i$ | Относительная частота $\frac{n_i}{n}$ | Эмпирическая частота $n_{x_i}$ | $F(x) = \frac{n_{x_i}}{n}$ |
|:-----:|:-------------:|:-------------------------------------:|:------------------------------:|:--------------------------:|
|   1   |      10       |             $\frac{1}{6}$             |               0                |             0              |
|   2   |       9       |            $\frac{3}{20}$             |               10               |       $\frac{1}{6}$        |
|   3   |       6       |            $\frac{1}{10}$             |               19               |      $\frac{19}{60}$       |
|   4   |      15       |             $\frac{1}{4}$             |               25               |       $\frac{5}{12}$       |
|   5   |       8       |            $\frac{2}{15}$             |               40               |       $\frac{2}{3}$        |
|   6   |      12       |             $\frac{1}{5}$             |               48               |       $\frac{4}{5}$        |

Получим следующую эмпирическую функцию распределения:

\begin{equation*}
F(x) = 
  \begin{cases}
    0, x \leq 1 \\
    \frac{1}{6}, x \in (1, 2] \\
    \frac{19}{60}, x \in (2, 3] \\
    \frac{5}{12}, x \in (3, 4] \\
    \frac{2}{3}, x \in (4, 5] \\
    \frac{4}{5}, x \in (5, 6] \\
    1, x > 6
  \end{cases}
\end{equation*}

```python
def F(x):
    if x <= 1:
        return 0
    if 1 < x <= 2:
        return 1./6
    if 2 < x <= 3:
        return 19./60
    if 3 < x <= 4:
        return 5./12
    if 4 < x <= 5:
        return 2./3
    if 5 < x <= 6:
        return 4./5
    if x > 6:
        return 1

X = np.linspace(-1, 7, 10000)
Y = [y for y in map(F, X)]
plt.scatter(X, Y, marker='_')

plt.title('График эмпирической функции распределения')
plt.xlabel("Значения СВ $X$")
plt.ylabel("$F(x) = P(X < x)$")
plt.show()
```
